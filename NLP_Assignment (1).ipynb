# Default Model

```python
#Text_summarization / CV_summarization
summarizer = pipeline("summarization", )
outputs = summarizer(text, max_length=60, clean_up_tokenization_spaces=True)
print(outputs[0]['summary_text'])
```

```python
#Question_Answering
reader = pipeline("question-answering")
question = "What is the higest qualification?"
outputs = reader(question=question, context=text)
pd.DataFrame([outputs])
```

```python
#Machine_Translation
translator = pipeline("translation_en_to_de")
outputs = translator(text, clean_up_tokenization_spaces=True, min_length=100)
print(outputs[0]['translation_text'])
```

```python
#Named_Entity_Recognition
ner_tagger = pipeline("ner", aggregation_strategy="simple")
outputs = ner_tagger(text)
pd.DataFrame(outputs)
```

```python
#Zeroshot_classification_pipeline
zero_shot_classification = pipeline("zero-shot-classification")
outputs = zero_shot_classification(text, candidate_labels=["education", "work experience", "personal goals"])
pd.DataFrame(outputs)
```

# Model checkpoint https://huggingface.co/distilbert/distilbert-base-uncased-finetuned-sst-2-english

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch

# Load tokenizer and model
checkpoint = "distilbert/distilbert-base-uncased-finetuned-sst-2-english"
tokenizer = AutoTokenizer.from_pretrained(checkpoint)
model = AutoModelForSequenceClassification.from_pretrained(checkpoint)
```

```python
#Change the input to other input statements, and understand the tokenization, input_ids, and other inputs to the model.

inputs = tokenizer("K2, at 8,611 metres (28,251 ft) above sea level, is the second-highest mountain on Earth", return_tensors="pt")

print("Tokens:", inputs.tokens())
print("Input IDs:", inputs["input_ids"])
print("Attention Mask:", inputs["attention_mask"])
```

```python
#Check the input, input_ids, tokens, special tokens of a tokenizer, embedding dimension
print("Special tokens:", tokenizer.special_tokens_map)
print("All tokens (with special):", tokenizer.convert_ids_to_tokens(inputs["input_ids"][0]))
print("*"*50)
embedding_dim = model.base_model.config.hidden_size
print("Embedding dimension:", embedding_dim)
```

```python
#Check the output components, loss, logits, and output and transform logits to final softmax outputs
with torch.no_grad():
    logits = model(**inputs).logits

predicted_class_id = logits.argmax().item()
print("Prediction: " + model.config.id2label[predicted_class_id])
```
